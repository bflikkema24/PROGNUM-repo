### Julian Date, a algorithm with some pitfalls (4 pt) <font color='red'><b>HOMEWORK</b></font>

In the online article: [A one-line algorithm for Julian Date](http://adsabs.harvard.edu/full/1983IAPPP..13...16F)
we find an old and very compact formula for programmable calculators to calculate a Julian Date given a day, month and year.   
A Julian Date is in fact a
day number corresponding to the number of elapsed days since the beginning of a cycle of 7980 years.
That makes it a useful tool to calculate how many days there are between two dates.
The one-line algorithm is valid for the period that started in 1582 with the Gregorian calendar.
The formula from the article is:


``JD = 367*Y -7*(Y+(M+9)/12)/4 - 3*((Y+(M-9)/7)/100 + 1)/4 + (275*M)/9 + D + 1721029-0.5``

The article notes: **It is important to note that after all divisions in this expression only that portion of the
quotient to the left of the decimal point is retained and that the quotient is not rounded off.**

* Use your notebook to develop an implementation of this formula.
* Then create a Python script called ``JD.py`` (store it in your bin directory on Kapteyn) that prompts a user to enter a   day, month and year.  
  For this date, the script calculates the Julian Date (JD) and prints it. The day can be entered as a floating point number to allow for fractional days. 
* The article provides a test case: the Julian Date for January 4, 1700 is 2341975.5.    
  Prove that your code gives the same result.
* Extend the program so that it can calculate the number of days between two dates in the Gregorian calendar.
* What is your age expressed in days?

**Warning**:  Copy/Paste the above formula will also copy symbols that are unknown to Python.
      
**Answers**:

````python
D = float (input("insert your day"))
M = float (input("insert your month"))
Y = float (input("insert your year"))

JD = 367* Y - 7 * ( Y + (M+9)//12)//4 - 3 * ((Y + (M-9)//7)//100 + 1)//4 + (275 * M)//9 + D + (1721029 - 0.5)


D1 = float (input("insert your day1"))
M1= float (input("insert your month1"))
Y1 = float (input("insert your year1"))

JD1 = 367* Y1 - 7 * ( Y + (M1+9)//12)//4 - 3 * ((Y1 + (M1-9)//7)//100 + 1)//4 + (275 * M1)//9 + D1 + (1721029 - 0.5)

D2 = float (input("insert your day2"))
M2= float (input("insert your month2"))
Y2 = float (input("insert your year2"))

JD2 = 367* Y2 - 7 * ( Y + (M2+9)//12)//4 - 3 * ((Y2 + (M2-9)//7)//100 + 1)//4 + (275 * M2)//9 + D2 + (1721029 - 0.5)

JD3 = JD2 - JD1
print(f"You are {JD3} days old")
````

### Uniqueness in lists using the `set` object (2 pt) <font color='red'><b>HOMEWORK</b></font>

Given two lists with a number of gamma ray bursts per night for two different periods

```python
    night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
    night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
```

Use Python sequence type `set` to create a **list**:

1. with values that are both in `night1` and `night2`.
1. with values that are the difference between `night1` and `night2`.
1. with values that contains all items from both sets, duplicates are excluded.
1. after an update of `night1` by removing all its elements that are not available in `night2`.

**Answers**:

````python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
#1
intersection = set(night1) & set(night2)
print(intersection)
#2
difference = set(night1) - set(night2)
print(difference)
#3
Both_nights = night1 + night2
combined = set(Both_nights)
Values = [x for x in Both_nights if x in combined or combined(x)]
print(Values)
#4
updated_night1 = [x for x in night1 if x in night2]
result = set(updated_night1)
print(result)
````

### Apply NumPy functions to data (2pt) <font color='red'><b>HOMEWORK</b></font>

* Create a NumPy array `x` with 100 angles in degrees between 0 and 360 degrees using ``np.linspace()``
* Calculate the ``np.sinc(x)`` values of ``x``, with ``x`` in radians. The definition of the sinc function is:

  $$\mathrm{sinc}(x) = \frac{sin(\pi x)}{\pi x}$$

* Print the first 10 values of the result. 
  
**Answer**

````python
import numpy as np
x = np.linspace(0,360,100)
print(x)
x_radians = np.deg2rad(x)

sinc_values = np.sinc(x_radians / np.pi)
print(sinc_values[0:10])
````
