```python
from astropy.io import fits
import numpy as np
hdulist = fits.open('m101.fits')
hdulist.info()
# Select the first row for data
hdr = hdulist[0].header
for key in hdr[:5]:
    print(f"{key:18s} = {hdr[key]}")

dat = hdulist[0].data
print(f" The data is of shape:", type(dat))
dat.dtype

from scipy import stats
# Using numpy functions to determine the required statistical values.
flatdat = dat.flatten()
minimum = np.min(flatdat)
print(f" The minimum of the data is: {minimum}")
maximum = np.max(flatdat)
print(f" The maximum of the data is: {maximum}")
mean = np.mean(flatdat)
print(f" The mean of the data is: {mean}")
std = np.std(flatdat)
print(f" The standard deviation of the data is: {std}")
median = np.median(flatdat)
print(f" The median of the data is: {median}")
mode, counts = stats.mode(flatdat)
print(f" The mode of the data is {mode} with {counts} counts. ")
# Use the calculated values to set up the histogram, with the desired lines.
from matplotlib import pyplot as plt
from matplotlib import figure

plt.hist(flatdat, alpha = 0.5)
plt.axvline(mean, label = "Mean")
plt.axvline(median, color = "red", label = "Median")
plt.axvline(mode, color = "yellow", label = "Mode")
plt.legend()
plt.show()

# Converting the data into an np array
fig = plt.figure()
#fig, axes = plt.subplots(1,2)
frame = fig.add_subplot(1,2,1)
data_array = np.asarray(flatdat)
median_array = np.median(data_array)
cutoff = median_array
filter = np.where(data_array > cutoff)
image = frame.imshow(dat, vmin=median, origin="lower")
colorbar = fig.colorbar(image, ax = frame, fraction=0.045, pad = 0.04 )
frame.set_xlabel("CTYPE1")
frame.set_ylabel("CTYPE2")

plt.show()
new_header = hdr.copy()
new_header['Bunit'] = ('Counts')
hdulist.close()
```

```python
from astropy.io import fits
import numpy as np
from matplotlib import pyplot as plt
hdulist = fits.open('m101.fits')
hdulist.info()
dat = hdulist[0].data
plt.imshow(dat, origin="lower")
plt.show()
rectangle_data = dat[20:40, 10:30]
flat_rectangle_data = rectangle_data.flatten()
RMS = (np.mean(rectangle_data**2)**0.5)
print(f" The RMS is: {RMS}")
mean_rectangle = np.mean(rectangle_data)
std_rectangle = np.std(rectangle_data)
print(f" The mean of the recangular area is: {mean_rectangle}")
print(f" The standard deviation in the data from this recangular area is: {std_rectangle}")
```

```python
from astropy.io import fits
import numpy as np
from matplotlib import pyplot as plt
hdulist0 = fits.open('noisemap0.fits')
hdulist1 = fits.open('noisemap1.fits')
hdulist2 = fits.open('noisemap2.fits')
# Extracting the data out of the noise fits.
dat0 = hdulist0[0].data
dat1 = hdulist1[0].data
dat2 = hdulist2[0].data

# Use np.std to calculate the standard deviation in the fits.
std0 = np.std(dat0)
std1 = np.std(dat1)
std2 = np.std(dat2)
stack = np.stack([dat0, dat1, dat2], axis=0)

mean_stack = np.mean(stack, axis = 0)
std = np.std(mean_stack)
print(std)
plt.imshow(mean_stack, origin = "lower")
plt.show()
```

## 6.6

```python
from astropy.io import fits
from astropy.table import Table
import numpy as np


# Open the FITS file
hdul = fits.open('APOGEE_stars.fits')

# Access the primary data (in this case, it is a Table)
data = hdul[1].data  # Often, the first extension contains the table
colnames = hdul[1].columns.names # List of the column names
# The biggest difference is the index while reading an image and a table. For an image you use the index 0 and for a table the index 1.
# The image gives the data directly, while the table gives the data dor each column

# Converting file to a table
table = Table(data)

# The flag for our new column
flag = ((data['STARFLAG'] == 0) & (data['ASPCAPFLAG'] == 0) & (data['SNR'] > 10) & (data['GAIAEDR3_PARALLAX'] / data['GAIAEDR3_PARALLAX_ERROR'] > 5) &
(~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) & (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) & (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG'])))  # ~ is for saying that something is not something

# Adding column to table
dataflag = flag.astype(int)  # Transforms True to 1 and False to 0
table['DATAFLAG'] = dataflag  # Adds the column DATAFLAG and looks whether the value should be 1 or 0

# Accessing the apparent magnitude table
m = table['GAIAEDR3_PHOT_G_MEAN_MAG']
p = table['GAIAEDR3_PARALLAX']

# Calculating the absolute magnitude
valid_p = p > 0  # Creating a mask
abs_mag = np.full(len(p), -1.0)  # Creates a full array of the length of p filled with -1.0
abs_mag[valid_p] = m[valid_p] - 5 * np.log10(p[valid_p]/1000) + 5  # Calculating the absolute magnitude for only the values that the absolute magnitude can be calculated
table['ABSOLUTE_MAG'] = np.where(table['DATAFLAG'] == 1, abs_mag, -1.0)  # Adds the values for the absoulte magnitude to the table, only when dataflag is 1, else it puts -1.0

# Creating a FITS file called APOGEE_update.fits
table.write('APOGEE_update.fits', format = 'fits', overwrite = True)  # Creating a new fits file called APOGEE_update.fits

# Printing the newly created column
print(table['ABSOLUTE_MAG'])
print(table['DATAFLAG'])
```

## 6.10

```python
from astroquery.gaia import Gaia

query = """
SELECT source_id, ra, dec, parallax, parallax_error, phot_g_mean_mag, phot_bp_mean_mag, phot_rp_mean_mag 
FROM gaiadr3.gaia_source 
WHERE parallax/parallax_error > 5
AND phot_g_mean_mag IS NOT NULL
AND phot_bp_mean_mag IS NOT NULL
AND phot_rp_mean_mag IS NOT NULL
"""
data = Gaia.launch_job(query)
results = data.get_results()

dataframe = results.to_pandas()
print(dataframe)
```
