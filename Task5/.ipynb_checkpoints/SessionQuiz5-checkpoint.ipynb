```python
import numpy as np
# Input variable that will ask the user for the filename.
filename = input("Enter your filename: ")
f = np.loadtxt(filename)
# Assign x to the first column and y to the second column.
x = f[:,0]  
y = f[:,1]
print(x)
print(y)
# Use np functions to find the minimums, maximums and means
x_min = np.min(x)
x_max = np.max(x)
x_mean = np.mean(x)
print(f" The mean of x is {x_mean}, the minimum of x is {x_min} and the maximum of x is {x_max}.")
y_min = np.min(y)
y_max = np.max(y)
y_mean = np.mean(y)
print(f" The mean of y is {y_mean}, the minimum of x is {y_min} and the maximum of y is {y_max}.")
y_med = np.median(y)
# Set up a variable that will calculate the values in y2.
y2 = (y-y_med)**2
print(y2)
```

```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = np.concatenate((a,b))
print(result)
vorm = result.shape
print(vorm)
```

The difference in shape between the sets of arrays, is that there will not only be the amount of columns, so 6 in the first case, but also a number of rows.

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.concatenate((a,b))
print(result)
result1 = np.concatenate((a,b), axis=0)
print(result1)
result2 = np.concatenate((a,b), axis=1)
print(result2)
```

The 'axis=0' part will not do anything different from the regular concatenate function. Adding 'axis=1' will make sure there is only two rows, resulting in have four columns.

```python
import imageio
import numpy as np

from matplotlib import pyplot as plt
from matplotlib.pyplot import figure, show
a = imageio.imread('m101BW.jpg')
plt.imshow(a, 'jet')
plt.show()
# Put the contours in the first plot
plt.contour(a, levels = 100)
plt.show()
# Set up a variable where 
cutoff = np.percentile(a, 90)
top_10 = np.where(a>= cutoff, a, 0)
plt.imshow(top_10)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from imageio.v2 import imwrite, imread 
from matplotlib.pyplot import figure, show, imshow
filename = "ngc6946.txt"
# Use np.loadtxt to put the numbers in the txt file into an array
data = np.loadtxt(filename)
#print(data)
data_shape = data.reshape(40,50,70)
first_plane = data_shape[0,:,:]

plane = plt.imshow(first_plane, "jet", origin = "lower")
plt.colorbar(plane)

# Calculating the mean and std
contour1 = np.mean(first_plane)
std = np.std(first_plane)
contour2 = contour1 + std
contour3 = contour2 + std
plt.contour(first_plane, levels = [contour1, contour2, contour3])
plt.show()
```

## 5.12

```python
from scipy.constants import c
class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self):
        z = self.redshift()
        v = c * z
        v_in_km = v/1000
        return v_in_km
gal2 = Galaxy("NGC1832", 3994.7)
print(f" the velocity of {gal2.name} is {gal2.velocity()} km/s.")
```

## 5.13

```python
import numpy as np
from matplotlib.pyplot import figure, show, imshow
from matplotlib import pyplot as plt
Nx = 400
Ny = 300
data = np.random.random((Ny,Nx))
plt.show()
fig = figure()
frame = fig.add_subplot(1,1,1, aspect=1)

from matplotlib.patches import Ellipse

class AstroEllipse(Ellipse):
     def __init__(self, xc, yc, semi_major, semi_minor, position_angle, **args):
        # We use the (sort of) astronomical angle where 'angle' defines)
        # the position angle of the major axis of a ring projected onto the sky
        # (i.e. an ellipse). This is an angle taken in anti-clockwise direction between
        # the north direction on the sky and the major axis of the receding half of the
        # galaxy (Rots 1975, astron, astrophys 45, 43).
        # Here we assume (!) north is in the positive Y direction.
        # If not, you need to correct the angle for the rotation of the image.
        # Parameter 'position_angle' enters in degrees
        assert(semi_major >= semi_minor)  # Prerequisite to make angle unambiguous
        XY = (xc, yc)                  # MPL requires center as a tuple
        angle_mpl = position_angle     # Angle for MatPlotLib's ellipse patch
        height = 2.0*semi_major        # height: is along vertical axis for an unrotated ellipse
        width  = 2.0*semi_minor        # width:  is along horizontal axis for an unrotated ellipse
        super().__init__(XY, width, height, angle=angle_mpl, **args)

xcenter = 200
ycenter = 150
semimajor = 100;
semiminor= 50
theta = 30.0    # degrees, anti clockwise
ell = AstroEllipse(xcenter, ycenter, semimajor, semiminor, theta, fill=False, hatch='/')
frame.add_patch(ell)
frame.set_xlim(0,400)    # Limits are not automatically adjusted when patches are used
frame.set_ylim(0,300)
frame.grid()
imshow(data,interpolation='none', origin="lower", cmap='gray')
show()
```
