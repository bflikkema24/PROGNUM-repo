## Great circle distance between two locations on Earth (5pt)

Let $\lambda_1$ , $\phi_1$ and $\lambda_{2}$, $\phi_{2}$ be the geographical longitude and latitude of two 
observatories and $\Delta\lambda$, $\Delta\phi$ be their absolute differences; then the central angle on a sphere between them ($\Delta \sigma$ is given by the spherical law of cosines if one of the poles is used as an auxiliary third point on the sphere:

$$\Delta \sigma = \arccos {\bigl (}\sin \phi_{1}\sin \phi_{2}+\cos \phi_{1}\cos \phi_{2}\cos(\Delta \lambda ){\bigr )}$$

The $\arccos()$ function is provided by the math module as `acos()`. When the central angle is found (in radians), the actual arc length $d$ on a sphere of radius $r$ can be computed as

$$d=r\,\Delta \sigma$$

with $r$, the radius of the earth being 6371 km.

Write Python code to calculate the great circle distance in km between the observatories:

* Groningen Observatory: latitude: 53.240208 | Longitude: 6.53646
* Observatorio del Roque de los Muchachos, La Palma, Spain: latitude: 28.757723 | longitude: -17.885062

Round the distance to 2 decimal places using an f-string.

**Answer**:

```python
from math import *
x = 6.53646 - -17.885062
print(x)
inside_arccos =  sin(53.230208) * sin(28.75723) + cos(53.230208) * cos(28.757723) * cos(x)
delta_sigma = acos(inside_arccos)
print(delta_sigma)
d = 6371 * delta_sigma
print(f" The great circle distance is {d:.3} km")
```
