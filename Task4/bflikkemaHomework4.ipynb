###  Finding best fit model parameters (2 pt) <font color='red'><b>HOMEWORK</b></font>

A third step in our data exploration is to model the data. The correlation coefficient shows that we should be confident 
in using a straight line as a *model*. We assume that a Big Bang took place so at $D=0$ the velocity is 0.
This reduces the model to a straight line through the origin. An expression to find the best fit 
parameter for the slope of the line in model $V=H_0 D$ (or $Y = b X$) is to minimize residuals 
with respect to parameter $b$. We derive an expression for $b$ as follows:

$$\chi^2 = \sum_{i=0}^{N-1} {\left( y_i-bx_i \right)}^2$$

$$0 = \frac{\partial\chi^2}{\partial b} = -2\sum_{i=0}^{N-1}{x_i(y_i-bx_i)}$$

$$b = \frac{\sum {x_i y_i}}{{\sum {x_i^2}}}$$
