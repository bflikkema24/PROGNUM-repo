```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
# Setup the histogram, where 185 is the average length, 8 is the standard deviation and 100000 is the amount of measurements.
lengths = np.random.normal(185, 8, 100000)
average = np.mean(lengths)
standard_deviation = np.std(lengths)
plt.hist(lengths, bins = 40)
plt.axvline(average, color = 'red', label = "Average")          # Adds the vertical line
plt.axvspan(average - standard_deviation, average + standard_deviation, color = 'yellow', label = "Standard Deviation", alpha = 0.3)      # Adds the vertical shading area
plt.legend()
plt.show()

print (average)
print(standard_deviation)
# Set up a variable that will be used to find values of height smaller than 185.
x = np.linspace(0,185,100000)
s = scipy.stats.norm.cdf(x, average, standard_deviation)
prob = np.sum(s)
# Sum to add all probabilities from the singular lengths.
print(prob/10000)

# Two variables to calculate both the chances of being within the lowerrange of the std of the mean, and the upperrange of the std.
under_mean = scipy.stats.norm.cdf(average - 8, 185, 8)
upper_mean = scipy.stats.norm.cdf(average + 8, 185, 8)
prob_within_std = np.sum(under_mean) + np.sum(upper_mean)
print(prob_within_std)
```

```python
import numpy as np
def funcfilter(func, seq):
    """ This function is used to filter out values that will not work for the log function."""
    filtered = np.asarray(seq)
    a = filtered[filtered>0] # Using a Boolean array to filter the values that are lower than or equal to zero.
    function = func(a)
    return function
   
sq = np.array([-100, -10, 0, 10, 100])


result = funcfilter(np.log10, sq)
print("Result log10:", result)

        
        


```

## 4.16

```python
from astropy.cosmology import Planck18
import scipy
from scipy.integrate import quad
H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0
def t_L (z):
    """ This function is used to calculate the integral involved in the lookback time."""
    H = H0 * (Om0 * (1 + z)**3 + Ode0)**0.5      # Variable for H
    t_l = 1 / ((1 + z) * H)                      # Variable for the formula that will be integrated. 
  
    return t_l

# Calculate the integral giving the CLT.
CLT, error = quad(t_L,0,2)

a = CLT * 977.8
print(f"{a} Gyr.")

    
    
    
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde
x = np.random.normal(0, 1, 100000)
y = np.random.normal(2, 1, 100000)
# Calculating the probability density functions
x_pdf = gaussian_kde(x)
y_pdf = gaussian_kde(y)
# Create a range in which the probability density function should be evaluated.
data_points = np.linspace(-4, 6, 1000)
# Plotting of the histograms
plt.figure(figsize = (12,10))
plt.hist(x, color = "blue", bins = 30, edgecolor= "black", alpha = 0.5, density = True, label = 'X')
plt.hist(y, color = "red" , bins = 30, edgecolor = "black",alpha = 0.5, density = True, label = 'Y')
# Plotting of the lines from the probability density function.
plt.plot(data_points, x_pdf(data_points), color = "black", ls = 'dotted', linewidth = 4, label = 'X(PDF)')
plt.plot(data_points, y_pdf(data_points), color = "black", ls = 'dashed', linewidth = 4, label = 'Y(PDF)')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.xticks(np.arange(-6,9,1))
plt.annotate(r'$\mathcal{N}(0, 1)$',xy=(-1.8, 0.15), xytext=(-5.5, 0.145),
             arrowprops=dict(facecolor='blue', shrink=0.05), color='blue', fontsize=15)
plt.annotate(r'$\mathcal{N}(2,1)$',xy=(4, 0.15), xytext=(6, 0.145),
             arrowprops=dict(facecolor='r', shrink=0.05), color='r', fontsize=15)
plt.title( '1D Histogram + PDF') 
plt.grid(True, ls = '--', axis = 'y')
plt.legend()
plt.show()
```

```python
#2D Histogram
X = np.random.normal(0,1,100000)
Y = np.random.normal(2,1,100000)
bins = 30
range_lim = [[-17, 17],[-17,17]]

H, xedges, yedges = np.histogram2d(X, Y, bins=100, range = range_lim)

# Bin centers (needed for contour)
xcenters = (xedges[:-1] + xedges[1:]) / 2
ycenters = (yedges[:-1] + yedges[1:]) / 2

# Meshgrid for contour
Xgrid, Ygrid = np.meshgrid(xcenters, ycenters)

# 1. Flatten and sort the histogram values in descending order
H_flat = H.flatten()
H_sort = np.sort(H_flat)[::-1]

# 2. Calculate the cumulative sum of the sorted bins
cumulative_sum = np.cumsum(H_sort)
cumulative_prob = cumulative_sum / cumulative_sum[-1]

target_probs = [0.68, 0.95, 0.997]
levels = []

for p in target_probs:
    # Find the threshold bin value
    idx = np.where(cumulative_prob >= p)[0][0]
    levels.append(H_sort[idx])

levels = sorted(levels)

plt.figure(figsize=(8, 7))
plt.hist2d(X,Y, bins = bins, cmap = "Blues")
plt.colorbar(label='')
plt.grid(True, linestyle='--', color='black', alpha=0.7)


contours = plt.contour(
    Xgrid, Ygrid, H.T,
    levels=levels,
    colors="red",
    linewidths=2
)

fmt = {
    levels[2]: r'$1\sigma$',
    levels[1]: r'$2\sigma$',
    levels[0]: r'$3\sigma$'
}

plt.clabel(contours, contours.levels, inline=True, fmt=fmt, fontsize=12, colors='red')

plt.title(f"2D Histogram + Contour")


plt.show()
```
