```python
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
N = len(masses)
index = range(N)
escape_velocities =  []
for i in index:
            escape_velocity = (float((2 * G * masses[i])/radii[i]) ** 0.5)
            escape_velocities.append(escape_velocity) # Adds the calculated escape velocities to a list.
            print (f"The escape velocity of {names[i]} is {escape_velocity:.2e}")
            
print (escape_velocities)
escape_v_dict = dict(zip( names, escape_velocities)) # Zip combines the pairs of names and escape velocities, dict makes it into a dictionary.
print (escape_v_dict)
```
